```python
#!/usr/bin/env python
# coding: utf-8

import os 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
```

```python
# In[9]:
Carpeta = "C:/Users/Aguirre/data/_TESS_lightcurves_median_after_detrended" #directorio
for subdirectorio in os.listdir(Carpeta):
    print(subdirectorio)
    
    plot_dir = '_TESS_lightcurves_median_after_detrended_plots/'+subdirectorio
    
    if not os.path.exists(plot_dir):
        os.makedirs(plot_dir)
		
		
    
    for archivo in os.listdir(Carpeta+'/'+subdirectorio):
        if archivo.endswith(".lc"):
            print(archivo)
            
           
            # making data frame from csv file 
            data_raw = pd.read_csv(Carpeta+'/'+subdirectorio+'/'+archivo, delimiter=' ', na_values='*********', 
					usecols=[0,2,3], names = ['time', 'mag', 'magerr']) 
				
            # sorting by first name 
            data_raw.sort_values('time', inplace = True) 
				
            # dropping ALL duplicate values 
            data_raw.drop_duplicates(subset ="time", keep = False, inplace = True) 
				
            data_raw = data_raw[data_raw['mag'].notna()]	

            tess_lc_raw = data_raw.to_records(index=False)
	
            
            
            plt.clf()
            
            ax = plt.gca()
            ax.invert_yaxis()
            
            plt.title(archivo)
            plt.xlabel('Julian Date')
            plt.ylabel('mag clean')
            plt.ticklabel_format(axis = 'both' , style = 'plain')

            markers, caps, bars = plt.errorbar(tess_lc_raw['time'],tess_lc_raw['mag'],yerr=tess_lc_raw['magerr'], fmt='o', ecolor='black', capsize = 2, capthick = 2, markersize=1, color='k')
            [bar.set_alpha(0.1) for bar in bars]
            [cap.set_alpha(0.1) for cap in caps]
     

            plt.savefig(plot_dir+'/'+archivo+'.pdf',format="pdf", bbox_inches="tight")
```
