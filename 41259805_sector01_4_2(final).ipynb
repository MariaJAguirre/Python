```python
import matplotlib.pyplot as plt
```

```python
x = []
y = []
yerror = []
```

```python
for columnas in open('DatosPresentacion.txt','r'):
    values = [float(i) for i in columnas.split()]
    x.append(values[0])
    y.append(values[1])
    yerror.append(values[3])
    
```

```python
ax = plt.gca()
ax.invert_yaxis()

plt.title("238869272_sector01_3_2")
plt.xlabel('Julian Date')
plt.ylabel('mag clean')
plt.ticklabel_format(axis = 'both' , style = 'plain')

markers, caps, bars = plt.errorbar(x,y,yerr=yerror, fmt='o', ecolor='black', capsize = 2, capthick = 2, markersize=1, color='k')
[bar.set_alpha(0.1) for bar in bars]
[cap.set_alpha(0.1) for cap in caps]

plt.savefig("238869272_sector01_3_2.pdf",format="pdf", bbox_inches="tight")
plt.show()
```
