```python
data_folder = "data/_TESS_lightcurves_median_after_detrended"
```

```python
Carpeta = data_folder 

for subdirectorio in os.listdir(Carpeta):
    print(subdirectorio)

   
    for archivo in os.listdir(Carpeta+'/'+subdirectorio):
        if archivo.endswith(".lc"):
            print(archivo)
            
            data = pd.read_csv(Carpeta+'/'+subdirectorio+'/'+archivo, delimiter=' ', na_values='*********', usecols=[0,2,3], names = ['time', 'magnitude', 'error']) 
            data.sort_values('time', inplace = True) 
            data.drop_duplicates(subset ="time", keep = False, inplace = True) 
            data = data[data['magnitude'].notna()]
            tess_lc_raw = data.to_records(index=False)
            
            
            lc = [data['time'], data['magnitude'], data['error']]
            
            a = data['magnitude']

            fs = fts.FeatureSpace(data = ['time', 'magnitude', 'error'], exclude = ["AndersonDarling","StetsonK"])
            features, values = fs.extract(*lc)
            
        
            anderson_return = anderson(a)
            print('Anderson',anderson_return[0])
            
            avg = np.average(data['magnitude'])
            residual = (data['magnitude'] - avg)/data['error']
            stetson_k = np.sum(np.fabs(residual)) / np.sqrt(np.sum(residual * residual)) / np.sqrt(len(data['magnitude']))
            print('Stetson K',stetson_k)
            

            f.write("%s,%s" % (archivo,subdirectorio))
 
            for i in range(values.size):
                f.write(",%.8f"% (values[i]))
            
            f.write("%.8f,"%(anderson_return[0]))
            f.write("%.8f,"%(stetson_k))
            f.write("\n")
            f.flush()
            
    
f.close()            
```
